```python
# Чтобы получить информацию со всех страниц, делаем итерацию

a = 0
response = ["a"]
houses_list = []

while len(response) != 0:
  response = requests.get(f"https://www.anapioficeandfire.com/api/houses?page={a}&pageSize=10").json()
  a += 1
  houses_list += response

df = pd.DataFrame(houses_list)
df
```

```python
# Как я понял девизу соответствует параметр "hasWords", поэтому фильтрация сделана именно по нему, а не по указанному "hasTitles"

a = 0
response = ["a"]
houses_with_words_list = []

while len(response) != 0:
  response = requests.get(f"https://www.anapioficeandfire.com/api/houses?hasWords=True&page={a}").json()
  a += 1
  houses_with_words_list += response

df = pd.DataFrame(houses_with_words_list)
df
```
